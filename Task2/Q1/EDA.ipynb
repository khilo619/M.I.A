```python
#you can see some usless \n we can replace them with nan
results_df = results_df.replace('\\N', np.nan)
races_df = races_df.replace('\\N', np.nan)
status_df = status_df.replace('\\N', np.nan)
```

```python
results_df['milliseconds'] = pd.to_numeric(results_df['milliseconds'], errors='coerce')
results_df['fastestLap'] = pd.to_numeric(results_df['fastestLap'], errors='coerce')
results_df['rank'] = pd.to_numeric(results_df['rank'], errors='coerce')
results_df['fastestLapSpeed'] = pd.to_numeric(results_df['fastestLapSpeed'], errors='coerce')
results_df['position_numeric'] = pd.to_numeric(results_df['position'], errors='coerce')

# who Retired, finished the race, disqualified.....
results_df['is_dnf'] = results_df['positionText'].isin(['R', 'D', 'F', 'W'])
results_df['is_finished'] = results_df['statusId'] == 1
results_df['got_points'] = results_df['points'] > 0

#let's view the persentages of the flags we created
print(f"DNF rate: {results_df['is_dnf'].mean()*100:.2f} %")
print(f"Finishing rate: {results_df['is_finished'].mean()*100:.2f} %")
print(f"Points sccoring rate: {results_df['got_points'].mean()*100:.2f} %")
```

Merging in one Data with the most usefull features & Feature Engineering

```python
df_merged = results_df.merge(races_df[['raceId', 'year', 'name', 'date']], on='raceId', how='left')
df_merged = df_merged.merge(status_df, on='statusId', how='left')
df_merged.head(10)
```

```python
df_merged['positions_lost'] = df_merged['grid'] - df_merged['position_numeric']
#lower value ==> better final position

df_merged['decade'] = (df_merged['year'] // 10) * 10

eras = {
    1950: '1950s', 1960: '1960s', 1970: '1970s', 1980: '1980s', 1990: '1990s',
    2000: '2000s', 2010: '2010s', 2020: '2020s'
}
df_merged['era'] = df_merged['decade'].map(eras)

df_merged['performance_category'] = 'DNF'
df_merged.loc[df_merged['position_numeric'] <= 3, 'performance_category'] = 'TOP 3'
df_merged.loc[(df_merged['position_numeric'] < 3) & (df_merged['position_numeric'] >= 10), 'performance_category'] = 'TOP 10'
df_merged.loc[df_merged['position_numeric'] > 10, 'performance_category'] = 'No Points'

print(f"\nYear ranges from {df_merged['year'].min()} to {df_merged['year'].max()}")
print(f"\nEvery driver total race in this decade:\n\n{df_merged['era'].value_counts().sort_index()}")
print(f"\nPerformace Categories: {df_merged['performance_category'].value_counts()}")
```

```python
races_per_decade = df_merged.groupby('era')['raceId'].nunique().sort_index()
print(f"\nUnique races per decade:\n {races_per_decade.head(5)}")
```

```python
print("\nSample of positions lost/gained:")
sample = df_merged[['driverId', 'year', 'grid', 'position_numeric', 'positions_lost']].dropna().head(10)
print(sample)
```

```python
df_merged['positions_lost'] = df_merged['grid'] - df_merged['position_numeric']
#lower value ==> better final position

df_merged['decade'] = (df_merged['year'] // 10) * 10

eras = {
    1950: '1950s', 1960: '1960s', 1970: '1970s', 1980: '1980s', 1990: '1990s',
    2000: '2000s', 2010: '2010s', 2020: '2020s'
}
df_merged['era'] = df_merged['decade'].map(eras)

df_merged['performance_category'] = 'DNF'
df_merged.loc[df_merged['position_numeric'] <= 3, 'performance_category'] = 'TOP 3'
df_merged.loc[(df_merged['position_numeric'] < 3) & (df_merged['position_numeric'] >= 10), 'performance_category'] = 'TOP 10'
df_merged.loc[df_merged['position_numeric'] > 10, 'performance_category'] = 'No Points'

print(f"\nYear ranges from {df_merged['year'].min()} to {df_merged['year'].max()}")
print(f"\nEvery driver total race in this decade:\n\n{df_merged['era'].value_counts().sort_index()}")
print(f"\nPerformace Categories: {df_merged['performance_category'].value_counts()}")
```

```python
print("Who are the top 10 drivers by total points?\n")

driver_stats = df_merged.groupby('driverId').agg(
    {
        'points' : 'sum',
        'raceId' : 'count',
        'position_numeric' : ['count', 'mean'],
        'is_finished' : 'sum',
        'got_points' : 'sum',
        'year' : ['min', 'max']
    }
).round(2)

driver_stats.columns = ['total_points', 'total_races', 'races_finished', 'avg_position', 
                       'completed_races', 'points_finishes', 'career_start', 'career_end']


driver_stats['points_per_race'] = (driver_stats['total_points'] / driver_stats['total_races']).round(2)
driver_stats['completion_rate'] = (driver_stats['completed_races'] / driver_stats['total_races'] *100).round(2)
driver_stats['points_rate'] = (driver_stats['total_points'] / driver_stats['total_races'] *100).round(2)
driver_stats['career_years'] = driver_stats['career_end'] - driver_stats['career_start'] + 1


significant_drivers = driver_stats[driver_stats['total_races'] >= 100].copy()

print("Top 10 drivers by total points:")
print(significant_drivers.nlargest(10, 'total_points')[['total_points', 'total_races', 'points_per_race', 'career_start', 'career_end']])

print("\nTop 10 drivers by points per race (min 50 races):")
print(significant_drivers.nlargest(10, 'points_per_race')[['total_points', 'points_per_race', 'total_races', 'avg_position']])

print("\nTop 10 most reliable drivers (completion rate):")
print(significant_drivers.nlargest(10, 'completion_rate')[['completion_rate', 'total_races', 'points_per_race']])
```

```python
print("Which team scores the most points per race start?\n")

constructor_stats = df_merged.groupby('constructorId').agg({
    'points': 'sum',
    'raceId': 'count', 
    'position_numeric': ['count', 'mean'], 
    'is_finished': 'sum',
    'got_points': 'sum',
    'year': ['min', 'max']
}).round(2)

constructor_stats.columns = ['total_points', 'total_entries', 'races_finished', 'avg_position',
                           'completed_races', 'points_finishes', 'first_year', 'last_year']

constructor_stats['points_per_entry'] = (constructor_stats['total_points'] / constructor_stats['total_entries']).round(1)
constructor_stats['completion_rate'] = (constructor_stats['completed_races'] / constructor_stats['total_entries'] * 100).round(1)
constructor_stats['points_rate'] = (constructor_stats['points_finishes'] / constructor_stats['total_entries'] * 100).round(1)
constructor_stats['years_active'] = constructor_stats['last_year'] - constructor_stats['first_year'] + 1


significant_constructors = constructor_stats[constructor_stats['total_entries'] >= 100].copy()

print("Top 10 constructors by total points:")
print(significant_constructors.nlargest(10, 'total_points')[['total_points', 'total_entries', 'points_per_entry', 'first_year', 'last_year']])

print("\nTop 10 constructors by efficiency (points per entry, min 100 entries):")
print(significant_constructors.nlargest(10, 'points_per_entry')[['total_points', 'points_per_entry', 'total_entries', 'avg_position']])

print("\nTop 10 most reliable constructors (completion rate):")
print(significant_constructors.nlargest(10, 'completion_rate')[['completion_rate', 'total_entries', 'points_per_entry', 'years_active']])

print("\nTop 10 most successful constructors (points rate):")
print(significant_constructors.nlargest(10, 'points_rate')[['points_rate', 'points_per_entry', 'total_entries', 'avg_position']])
```

```python
print("Which season had the most DNFs, and what were the common causes?\n")

dnf_by_year = df_merged.groupby('year').agg(
    {
        'is_dnf': ['count', 'mean'],
        'raceId': ['nunique']
    }
).round(2)

dnf_by_year.columns = ['total_dnfs', 'dnf_rate', 'Races']
dnf_by_year['DNF_per_race'] = (dnf_by_year['total_dnfs'] / dnf_by_year['Races']).round(2)

print("TOP 10 SEASON WITH DNF'S:")
print(dnf_by_year.nlargest(10, 'total_dnfs'))
```

```python
print("Worst Performers: Who lost the most positions from start to finish?\n")

position_loses = df_merged.groupby('driverId').agg(
    {
        'positions_lost': ['mean', 'sum', 'count'],
        'raceId': 'count'
    }
).round(2)

position_loses.columns = ['Avg_Positions_Lost', 'Total_Positions_Lost', 'Races_With_Loss', 'Total_Races']
significant_losers = position_loses[position_loses['Races_With_Loss'] >= 100]


print("Top 10 drivers who lost most positions:")
# print(significant_losers.nlargest(10, 'Total_Positions_Lost'))
print(significant_losers['Total_Positions_Lost'].nlargest(10))
```

```python
print("Era Comparison: How do driver performances vary across different decades?\n")

era_comparison = df_merged.groupby('era').agg(
    {
        'points': 'mean',
        'is_dnf': 'mean',
        'got_points': 'mean',
        'positions_lost': 'mean'
    }
).round(2) * 100

era_comparison.columns = ['Avg_Points', 'DNF_Rate', 'Points_Rate', 'Avg_Positions_Lost']
print("Performance by era:")
print(era_comparison)
```
